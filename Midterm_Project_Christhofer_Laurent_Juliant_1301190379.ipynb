```python
Nasdaq_close= df_Nasdaq.drop(df_Nasdaq.columns[0:5], axis=1)
Nikkei_close= df_Nikkei.drop(df_Nasdaq.columns[0:5], axis=1) 
HangSeng_close= df_HangSeng.drop(df_HangSeng.columns[0:5], axis=1)
```

```python
df_Nasdaq2019["Day_Perc_Change"] = df_Nasdaq2019["Adj Close"].pct_change()*100
df_Nikkei2019["Day_Perc_Change"] = df_Nikkei2019["Adj Close"].pct_change()*100
df_HangSeng2019["Day_Perc_Change"] = df_HangSeng2019["Adj Close"].pct_change()*100
```

```python
df_Nasdaq2019.dropna(axis = 0, inplace = True)
df_Nikkei2019.dropna(axis = 0, inplace = True)
df_HangSeng2019.dropna(axis = 0, inplace = True)
```

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among Top Gainers"
    elif x > -7 and x <= -3:
        return "Among Top Losers"
    elif x > 7:
        return "Bull Run"
    elif x <= -7:
        return "Bear Drop"
```

```python
df_Nasdaq2019["Trend"]= df_Nasdaq2019["Day_Perc_Change"].apply(lambda x:trend(x))
df_Nasdaq2019.head()
```

```python
df_Nikkei2019["Trend"]= df_Nikkei2019["Day_Perc_Change"].apply(lambda x:trend(x))
df_Nikkei2019.head()
```

```python
df_HangSeng2019["Trend"]= df_HangSeng2019["Day_Perc_Change"].apply(lambda x:trend(x))
df_HangSeng2019.head()
```

```python
#NASDAQ

colors_list = ['orchid', 'palegreen', 'palevioletred', 'magenta', 'purple', 'firebrick']
Nasdaq_piechart = df_Nasdaq2019.groupby('Trend')
pie_label = sorted([i for i in df_Nasdaq2019.loc[:, 'Trend'].unique()])
plt.pie(Nasdaq_piechart['Trend'].count(), labels = pie_label, colors=colors_list, shadow=True,startangle=15,
        autopct = '%1.1f%%', radius = 3)
plt.show()
```

```python
#NIKKEI

colors_list = ['orchid', 'palegreen', 'palevioletred', 'magenta', 'purple', 'firebrick']
Nikkei_piechart = df_Nikkei2019.groupby('Trend')
pie_label = sorted([i for i in df_Nikkei2019.loc[:, 'Trend'].unique()])
plt.pie(Nikkei_piechart['Trend'].count(), labels = pie_label, colors=colors_list, shadow=True,startangle=15,
        autopct = '%1.1f%%', radius = 3)
plt.show()
```

```python
#HangSeng

colors_list = ['orchid', 'palegreen', 'palevioletred', 'magenta', 'purple', 'firebrick']
HangSeng_piechart = df_HangSeng2019.groupby('Trend')
pie_label = sorted([i for i in df_HangSeng2019.loc[:, 'Trend'].unique()])
plt.pie(HangSeng_piechart['Trend'].count(), labels = pie_label, colors=colors_list, shadow=True,startangle=15,
        autopct = '%1.1f%%', radius = 3)
plt.show()
```

### Soal 4 (30 poin)
Dapat disimpulkan dari Visualisasi diatas dengan acuan Daily Percentage Change (Piechart) dan Trend, bahwa masing-masing harga saham lebij dominan ke arah Slight or No Change, karena tidak adanya perubahan pada harga saham tersebut. Apabila dibuat plot maka akan membentuk suatu Trend mendatar (tidak naik dan tidak turun)

#### Code

```python
# load stock market data
df_Nasdaq = pd.read_csv("Nasdaq.csv") 
df_Nikkei = pd.read_csv("Nikkei.csv") 
df_HangSeng = pd.read_csv("Hang_Seng.csv") 

combined_df = pd.concat([df_Nasdaq["Adj Close"], df_Nikkei["Adj Close"], df_HangSeng["Adj Close"]], axis=1)
combined_df.columns = ["Nasdaq", "Nikkei", "HangSeng"]
combined_df.head()
```

```python
combined_df.dropna(inplace = True, axis = 0)
combined_df.head()
```

```python
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)
pct_chg_df.head()
```

```python
volatility = pct_chg_df[["Nasdaq", "Nikkei", "HangSeng"]].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 10))
```

#### Penjelasan 
Terlihat dari hasil volatility, dari semua harga saham harga ketiga stock saham menandakan trend sideways yang mana cenderung tidak naik dan tidak turun tetapi mendatar
